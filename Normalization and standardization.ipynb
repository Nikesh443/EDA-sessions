## Normalization using formula

```python
# step-1: calculate min value of P_wage=min_wage
# step-2: Calulate max value of p_wage
# step-3:  Dr=Max_wage-min_wage
# step-4: p_wage-min_wage
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path="C:\\Users\\pc\\Documents\\Visadataset.csv"
visa_df=pd.read_csv(file_path)
visa_df
```

```python
x=visa_df['prevailing_wage']
min_pwage=visa_df['prevailing_wage'].min()
max_pwage=visa_df['prevailing_wage'].max()
Mean=np.mean(visa_df['prevailing_wage'])
SD=np.std(visa_df['prevailing_wage'])
NR=x-min_pwage
DR=max_pwage-min_pwage
visa_df['X_normalized']=NR/DR
visa_df
```

```python
visa_df[['prevailing_wage','X_normalized']]
```

```python
max_id=visa_df['X_normalized'].idxmax()
min_id=visa_df['X_normalized'].idxmin()
max_id,min_id
```

```python
visa_df[['prevailing_wage','X_normalized']].iloc[[max_id,min_id]]
```

```python
file_path="C:\\Users\\pc\\Documents\\Visadataset.csv"
visa_df=pd.read_csv(file_path)
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
visa_df['N_p_wage']=mms.fit_transform(visa_df[['prevailing_wage']])
visa_df
```

## Z-score

```python
file_path="C:\\Users\\pc\\Documents\\Visadataset.csv"
visa_df=pd.read_csv(file_path)
X=visa_df['prevailing_wage']
Mean=np.mean(X)
STD=np.std(X)
Nr=X-Mean
Dr=STD
visa_df['Z_score_p_wage']=Nr/Dr
visa_df
```

```python
visa_df[['prevailing_wage','Z_score_p_wage']]
```

```python
plt.subplot(1,2,1)
plt.boxplot(X)
plt.subplot(1,2,2)
plt.boxplot(visa_df['Z_score_p_wage'])
```
