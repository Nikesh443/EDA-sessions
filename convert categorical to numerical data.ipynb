## one hot encoder

- It means at a time only one will be ON(1/True), others are OFF(0/False)

- suppose case_status has two unique lables

- certified and denied

- In one hot encoder it will create two new columns

- like below


|case_status|certified|denied|
|-----------|---------|------|
|certified|1|0|
|denied|0|1|

### Advantages 
    
- one hot encoder new columns are orthogonal each other

- orthogonality means 90 degrees phase shift

- 90 degree phase shift means perpendicular to each other

- Independent means no relation

- It is very important property to avoid relation between variables

## Draw back

- Assume that you have 100 unique lables for a column

- so it will create 100 new columns

- To process 100 columns we require more memory,more time, more hardware 

- This is called curse of dimensionality
